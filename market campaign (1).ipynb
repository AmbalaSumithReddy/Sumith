```python
import matplotlib.pyplot as plt
import seaborn as sns
cols = data.columns
colors = ['black' , 'blue']
sns.heatmap(data[cols].isnull(),cmap=sns.color_palette(colors),cbar= True)
```

```python

fig, axes = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 3]})

sns.boxplot(data=data["Income"], ax=axes[0], color='skyblue', width=0.5, orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Income Levels")
sns.histplot(data["Income"], kde=True, ax=axes[1], color='purple', bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Income Levels")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
